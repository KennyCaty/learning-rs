```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('grade_data.csv', encoding="GBK")

# 显示数据框的前几行
# print(df.head())

# 选择前四列
first_four_columns = df.iloc[:, :4]

# 选择第四列及之后表头包含“必修”的列
required_columns = df.loc[:, df.columns[3:][df.columns[3:].str.contains('必修')]]

# 合并两部分数据
result_df = pd.concat([first_four_columns, required_columns], axis=1)

# 删除最后两行
result_df = result_df.iloc[:-2]

# 删除第2、3和4列
result_df = result_df.drop(df.columns[[1, 2, 3]], axis=1)

# 删除数据值少于10个的列
result_df = result_df.dropna(thresh=10, axis=1)

# 将第一列数据设置为索引
result_df.set_index(result_df.columns[0], inplace=True)

# 删除原始的第一列
result_df.drop(columns=result_df.columns[0], inplace=True)

# 将索引名设置为空字符串
result_df.index.name = ''

# 显示结果
print(result_df.head())
```

```python
result_df.to_csv('grade_data_test.csv',encoding='GBK')
```
